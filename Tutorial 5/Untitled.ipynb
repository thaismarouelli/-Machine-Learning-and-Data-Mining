```python
%matplotlib inline 
import pandas as pd 
import numpy as np 
import seaborn as sns; sns.set(font_scale=1.2) 
import matplotlib.pyplot as plt 

data=pd.read_csv('svm.csv') 
```

```python
sns.lmplot(x='x1', y='x2',
           data=data,
           hue='r',
           palette='Set1', 
           fit_reg=False, 
           scatter_kws={"s":50})

plt.show()
```

```python
from sklearn import svm 
points = data[['x1','x2']].values
result = data['r']

clf = svm.SVC(kernel = 'linear')
clf.fit(points, result) 

print('Vector of weights (w) = ', clf.coef_[0])
print('b = ', clf.intercept_[0]) 
print('Indices of support vectors =', clf.support_)
print('Support vectors =', clf.support_vectors_)
print('Number of supporty vbectors for each class= ', clf.n_support_)
print('Coefficients of the support vector in the decision function =', np.abs(clf.dual_coef_))
```

## Plotting the Hyperplane and Margins 

```python
#w is the vector of weights 
w = clf.coef_[0]

#slope of the hyperplane 
slope = -w[0]/w[1]

b = clf.intercept_[0]

#find the coordinates for the hyperplane 
xx = np.linspace(0,4) 
yy = slope*xx-(b/w[1])

#plot the margin 
s=clf.support_vectors_[0]
yy_down = slope*xx+(s[1] - slope * s[0]) 

s=clf.support_vectors_[-1]
yy_up = slope * xx + (s[1] - slope * s[0])

#plt the points 
sns.lmplot(x='x1', y='x2', data=data, hue='r', palette='Set1', fit_reg=False, scatter_kwn={"s":70})

#plot the hyperplane  
sns.plot(xx, yy, linewindth=2, color='green');

#plot the 2 margin 
plt.plot(xx, yy_down, 'k--') 
plt.plot(xx, yy_up, 'k--')
```
